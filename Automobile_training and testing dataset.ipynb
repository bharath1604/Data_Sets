## Finding Missing values

```python
data.isna().sum()
```

### Missing Data

```python
data_temp=data[data['normalized-losses']!='?']
normalised_mean=data_temp['normalized-losses'].astype(int).mean()
data['normalized-losses']=data['normalized-losses'].replace('?',normalised_mean).astype(int)
data_temp=data[data['price']!='?']
normalised_mean=data_temp['price'].astype(int).mean()
data['price']=data['price'].replace('?',normalised_mean).astype(int)

data_temp=data[data['horsepower']!='?']
normalised_mean=data_temp['horsepower'].astype(int).mean()
data['horsepower']=data['horsepower'].replace("?",normalised_mean).astype(int)

data_temp=data[data['peak-rpm']!='?']
normalised_mean=data_temp['peak-rpm'].astype(int).mean()
data['peak-rpm']=data['peak-rpm'].replace('?',normalised_mean).astype(int)

data_temp=data[data['bore']!='?']
normalised_mean=data_temp['bore'].astype(float).mean()
data['bore']=data['bore'].replace('?',normalised_mean).astype(float)

data_temp=data[data['stroke']!='?']
normalised_mean=data_temp['stroke'].astype(float).mean()
data['stroke']=data['stroke'].replace('?',normalised_mean).astype(float)

data['num-of-doors']=data['num-of-doors'].replace('?','four')
data.head()
```
